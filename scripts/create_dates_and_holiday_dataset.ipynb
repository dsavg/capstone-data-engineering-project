```python
# import libraries
import numpy as np
import pandas as pd
import datetime as dt
import holidays
```

```python
start_date = dt.datetime(2022, 12, 1)
end_date = dt.datetime(2024, 1, 1)
# periods between start and end date
delta = end_date - start_date
```

```python
# create date dataframe by hour
date_df = pd.date_range(start_date, 
                        periods=delta.days, 
                        freq='D').to_frame(index=False, 
                                           name='date_utc')
```

```python
date_df.head()
```

```python
# add all date related fields
date_df['day'] = date_df.date_utc.dt.day
date_df['day_name'] = date_df.date_utc.dt.day_name()
date_df['week'] = date_df.date_utc.dt.isocalendar().week
date_df['weekday'] = date_df.date_utc.dt.weekday
date_df['month'] = date_df.date_utc.dt.month
date_df['month_name'] = date_df.date_utc.dt.month_name()
date_df['year'] = date_df.date_utc.dt.year
```

```python
date_df.head()
```

```python
# importing holidays of top 5 countries using reddit 
# https://www.similarweb.com/website/reddit.com/#geography
usa_holidays = holidays.US(years=date_df['year'].unique())
uk_holidays = holidays.UK(years=date_df['year'].unique())
ca_holidays = holidays.CA(years=date_df['year'].unique())
au_holidays = holidays.AU(years=date_df['year'].unique())
de_holidays = holidays.DE(years=date_df['year'].unique())
```

```python
date_df['united_state_holidays'] = date_df['date_utc'].map(usa_holidays)
date_df['is_united_state_holiday'] = (date_df['united_state_holidays']
                                      .apply(lambda x: False if x is np.NAN else True))

date_df['united_kingdom_holidays'] = date_df['date_utc'].map(uk_holidays)
date_df['is_united_kingdom_holiday'] = (date_df['united_kingdom_holidays']
                                        .apply(lambda x: False if x is np.NAN else True))

date_df['canada_holidays'] = date_df['date_utc'].map(ca_holidays)
date_df['is_canada_holiday'] = (date_df['canada_holidays']
                                .apply(lambda x: False if x is np.NAN else True))

date_df['australia_holidays'] = date_df['date_utc'].map(au_holidays)
date_df['is_australia_holiday'] = (date_df['australia_holidays']
                                   .apply(lambda x: False if x is np.NAN else True))

date_df['germany_holidays'] = date_df['date_utc'].map(de_holidays)
date_df['is_germany_holiday'] = (date_df['germany_holidays']
                                 .apply(lambda x: False if x is np.NAN else True))
```
